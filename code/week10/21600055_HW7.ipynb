```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x_data = [1.0,2.0,3.0] # Independent variable
y_data = [2.0,4.0,6.0] # Dependent variable
# y = 2x, w = 2 model
w = 0.0 # weight
def forward(x): # get y
    return x * w
    
def loss(y_pred,y): # difference between predicted value and actual value
    return (y_pred - y ) * (y_pred - y) # loss function (yn - f(x))2

for epoch in range(30): #find w and loss
    for x, y in zip(x_data,y_data):
        y_pred = forward(x) # get prediected y
        l = loss(y_pred,y)
        w = w + 0.05
    print("Epoch",epoch,"\t(x,y): ",x,y,"\tw=",'%.2f'%w,"\tloss=","%.2f"%l)
    plt.scatter(w,l,color="r")
```

```python
x_data = [1,2,3]
y_data = [2,4,6]

def gradient(x,y): # get gradient
    return 2 * x * (x * w - y)

w = 1.0 # initial weight
for epoch in range(30): # 30 training
    for x, y in zip(x_data,y_data):
        y_pred = forward(x) 
        l = loss(y_pred,y)
        grad = gradient(x,y) # get gradient of loss function. (wx(predicted) - y)^2 => 2x(wx -y)
        w = w - 0.005 * grad # learning process
    print("Epoch",epoch,"\t(x,y): ",x,y,"\tw=",'%.2f'%w,"\tloss=","%.2f"%l,"\tGrad=",'%.2f'%grad)
    
    plt.scatter(w,l,color="r")
```

```python
x_data = [1.0,2.0,3.0,4.0,6.0]
y_data = [2.0,4.0,6.0,8.0,12.0]
w = 0.0 # weight
b = 0.0 # bias, w0

n_data = len(x_data) 
epochs = 100
learning_rate = 0.01

for i in range(epochs): # training 100 epochs
    print("Process ", i+1)
    for x_i, y_i in zip(x_data,y_data):
        y_hat = x_i * w + b # predicted y = x1 * w1 + w0
        loss = ((y_hat - y_i)**2) / n_data
        grad_w = ((w * x_i - y_i + b)*2 * x_i) / n_data
        grad_b = ((w * x_i - y_i + b)*2) / n_data
        
        w -= learning_rate * grad_w # w(i+1) = wi - learning rate * grad_w(2x(wx-y))
        b -= learning_rate * grad_b # b(i+1) = bi - learning rate * grad_b(2x(wx-y))
        print("weight = ",w,"\t\t","bias = ",b)
    plt.scatter(w,loss,color="r")
```

# sklearn Regression

```python
df = pd.read_excel('data_set_train.xlsx') # read excel file
df.head()
```

```python
df_2017q1 = df[df['yyyyqrt(거래년도 분기별)']=='2017Q1'] # slice data which yyyyqrt is 2017Q1
df_2017q1.corr().head() # print corelation
```

```python
# area vs price
plt.figure(figsize=(12,5))
plt.scatter(df_2017q1['area(면적)'],df_2017q1['price(가격)'],alpha=0.1)
plt.xlabel('area')
plt.ylabel('price')
plt.show() # graph of area ans price
```

```python
from sklearn.model_selection import train_test_split

# split data into training set and test set 
X = np.array(df_2017q1['area(면적)']).reshape(-1,1) # reshape (780,1)
y = np.array(df_2017q1['price(가격)']).reshape(-1,1)
train_x, test_x, train_y, test_y = train_test_split(X,y,test_size = 0.10,random_state = 2) # split data for train data and test data
# split random seed 2
print("train_x shape = ",train_x.shape,"train_y shape = ", train_y.shape)
print("test_x shape = ",test_x.shape,"test_y shape = ",test_y.shape)
```

```python
# Training
simple_lin_reg = LinearRegression(normalize = True)
simple_lin_reg.fit(train_x,train_y)
print("Fitting results: ","b = ", simple_lin_reg.intercept_, "w = ",simple_lin_reg.coef_)

# Prediction
test_y_pred = simple_lin_reg.predict(test_x)
print("Testing results: ","score = ",simple_lin_reg.score(test_x, test_y))

# Visulization
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(train_x,train_y,color='k')
plt.scatter(train_x,simple_lin_reg.predict(train_x),color="c")
plt.title('Training data')
plt.xlim(0,260)
plt.ylim(0,550000)

plt.subplot(1,2,2)
plt.scatter(test_x,test_y,color='k')
plt.scatter(test_x,test_y_pred,color='r')
plt.title("Testing data")
plt.xlim(0,260)
plt.ylim(0,550000)
```

# Multiple Regression

```python
df_2017q1_mf = df_2017q1.copy()
col = df_2017q1.columns

for c in col:
    if(df_2017q1_mf[c].dtypes == 'object'): # if column is string value, remove
        df_2017q1_mf = df_2017q1_mf.drop(c,axis=1)
        continue
    if(df_2017q1_mf[c].var() == 0): # if column is constant, remove
        df_2017q1_mf = df_2017q1_mf.drop(c,axis=1)
# if nan exist, replace it to mean value
df_2017q1_mf['Yongpae(용적률)'] = df_2017q1_mf['Yongpae(용적률)'].replace(np.nan,df_2017q1_mf['Yongpae(용적률)'].mean())
df_2017q1_mf['Gunpae(건폐율)'] = df_2017q1_mf['Gunpae(건폐율)'].replace(np.nan,df_2017q1_mf['Gunpae(건폐율)'].mean())
df_2017q1_mf
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_2017q1_mf.iloc[:,1:],\
                                                   df_2017q1_mf.iloc[:,0],\
                                                   test_size=0.10, random_state=2) # split train data and test data

print("X_train.shape = ",X_train.shape)
print("y_train.shape = ",y_train.shape)

print("X_test.shape = ",X_test.shape)
print("y_test.shape = ",y_test.shape)
```

```python
mul_reg = LinearRegression(normalize = True)
mul_reg.fit(X_train,y_train)

print("coef_: ",mul_reg.coef_) # print coefficients w
print("intercept_: ",mul_reg.intercept_)
```

```python
# evaluation
print("Training set score: {:.2f}".format(mul_reg.score(X_train,y_train)))
print("Test set score: {:.2f}".format(mul_reg.score(X_test,y_test)))
```

```python
from sklearn.linear_model import Lasso
#evaluation
lasso = Lasso(alpha = 1, normalize = True).fit(X_train,y_train)
print("Training score: {:.2f}".format(lasso.score(X_train,y_train))) # print train score
print("Test score: {:.2f}".format(lasso.score(X_test, y_test))) # print test score
print("Number of features used: ",np.sum(lasso.coef_!=0)) # sum if coefficient is not 0
```

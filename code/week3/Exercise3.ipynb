2. input: amazon_train_df.csv file

3. output: print each review's most frequent 5 words.

```python
import pandas as pd
import numpy as np
import nltk
from bs4 import BeautifulSoup
import string
from nltk.corpus import stopwords
import spacy
nlp = spacy.load('en_core_web_sm')
stop_words = set(stopwords.words('english'))
```

```python
train_df = pd.read_csv("amazon_train_df.csv") # read csv file
train_df
```

```python
train_df =train_df["4"] #slice only review
train_df
```

```python
def remove_punctuation(text): # function that remove punctuation like @,!, ? ... ""
    sent = []
    for t in text.split(" "): # split each word
        no_punct = "".join([c for c in t if c not in string.punctuation]) # if punctuation cut c
        sent.append(no_punct)
    sentence = " ".join(s for s in sent)
    return sentence
```

```python
def token_lemma(review): # functuon that lemma and tokenize
    doc = nlp(review.strip())
    tok_lem_sentence = [token.lemma_ for token in doc] # lemmatization
    return tok_lem_sentence
```

```python
def count_frequency(review): # function that return most 5 frequency words in dictionary type
    review = remove_punctuation(review)
    review = review.lower() # change all words to lowercase
    review = token_lemma(review)
    review = [w for w in review if not w in stop_words] # remove stop_words: a, an , the
    dictionary = {}
    for word in review: # word is non stop word
        if word not in dictionary: # make frequent dictionary
            dictionary[word] = 0
        dictionary[word] += 1
    dictionary = sorted(dictionary.items(), key = lambda x:x[1],reverse = True) # sort frequent value from max to min
    word_to_index = {}
    i = 0
    for (word, frequency) in dictionary: # save most 5 frequent words key, value
        if i > 4 :
            break
        else:
            i += 1
            word_to_index[i] = word
    return word_to_index
```

```python
for i in range(0,len(train_df)): # loop and print each review's most 5 frequent words.
    print("review",i+1,":",count_frequency(train_df.loc[i]))
```

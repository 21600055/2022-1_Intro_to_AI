1. Objective: Process the Mnist data lab

2. Input: mnist.pkl file

3. Output: image and label

```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd

print("loading data")
with open("./data/mnist.pkl","rb") as f:
    train_set, valid_set, test_set = pickle.load(f,encoding="latin1") # load mnist file
    
train_x, train_y = train_set # return each number's pixel brightness array and label array as tuple
test_x, test_y = test_set

train_x = pd.DataFrame(train_x) # make pixel brightness to data frame
train_y = pd.DataFrame(train_y, columns = ['label']) # make label array to dataframe
test_x = pd.DataFrame(test_x)
test_y = pd.DataFrame(test_y, columns = ["label"])

train_data = pd.concat([train_x,train_y],axis = 1) # concatenate label to x
test_data = pd.concat([test_x,test_y],axis = 1)

print(train_data.shape,test_data.shape)
train_data.head() # print first 5rows
```

```python
subset_image_x = train_data.iloc[:10,:-1] # data 10 rows all column except label
subset_image_y = train_data.iloc[:10,-1]# data 10 rows and last column label

for i, row in subset_image_x.iterrows(): # loop 10 rows
    ax = plt.subplot(2,5,i+1) # sub plot
    pixels = row.values.reshape((28,28)) # reshape each image to 28 X 28 size
    plt.imshow(pixels,cmap = "gray") # image show in grayscale
    plt.title("label: {}".format(subset_image_y[i]))
    
    plt.xticks([]) #erase ticks
    plt.yticks([])
```

# 1. get Corona excel data

```python
import pandas as pd

corona_df = pd.read_excel("corona.xlsx")
corona_df = corona_df.sort_values("서울시 기준일")
corona_df # training data
```

```python
for i in range(0,len(corona_df)):
    corona_df.loc[i,"서울시 기준일"] = corona_df.loc[i,"서울시 기준일"][0:10].replace(".","-")
```

```python
corona_df.drop(columns=["서울시 기준일"], inplace=True)
corona_df
```

# 2. Read weather excel data

```python
weather_df = pd.read_excel("weather.xlsx")
day = (weather_df.일시>="2020-02-05")
weather_df = weather_df[day][:]
weather_df # training data
```

```python
weather = weather_df[["일시","평균기온(°C)","평균 상대습도(%)"]]
weather = weather.reset_index()
weather.drop(columns=["index"], inplace=True)
weather
```

# 3. Linear Regression

```python
data = pd.concat([weather,corona_df],axis=1)
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
import numpy as np

x = np.array(data["평균기온(°C)"]).reshape(-1,1)
y = np.array(data["평균 상대습도(%)"]).reshape(-1,1)
z = np.array(data["증감율"]).reshape(-1,1)

train_x,test_x, train_y,test_y, train_z,test_z = train_test_split(x,y,z,test_size = 0.10,random_state=2)

print("train_x shape = ",train_x.shape,"train_y shape = ",train_y.shape,"train_z shape = ",train_z.shape)
print("test_x shape = ",test_x.shape, "test_y shape = ",test_y.shape,"test_z shape = ",test_z.shape)
```

```python
train_x = np.concatenate(train_x).tolist() # make 2 dimension => 1 dimension
train_y = np.concatenate(train_y).tolist()
train_z = np.concatenate(train_z).tolist()

X = {"평균기온(°C)" : train_x,
     "평균 상대습도(%)" : train_y
    }
X = pd.DataFrame(X)
X
```

```python
Y = {"증감율" : train_z }
Y = pd.DataFrame(data=Y)
Y
```

```python
def z_score(df): # normalization
    # copy the dataframe
    df_std = df.copy()
    # apply the z-score method
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std
    
# call the z_score function
X = z_score(X)

X
```

```python
test_x = np.concatenate(test_x).tolist() # make 2 dimension => 1 dimension
test_y = np.concatenate(test_y).tolist()
test_z = np.concatenate(test_z).tolist()

test_X = {"평균기온(°C)" : test_x,
     "평균 상대습도(%)" : test_y
    }
test_X = pd.DataFrame(test_X)
test_X
```

```python
test_Y = {"증감율":test_z}
test_Y = pd.DataFrame(test_Y)
test_Y
```

```python
regr = linear_model.LinearRegression()
regr = regr.fit(X, Y)
```

```python
pre = [] # predicted value list
for i in range(0,len(test_x)):
    predicted = regr.predict([[test_x[i],test_y[i]]])
    pre.append(predicted[0][0])
    error_rate  = (abs(test_z[i] - predicted) / test_z[i]) # get error_rate
    print("#",i+1,"predicted : ",predicted,"Real data: ",test_z[i],"Error rate: ",error_rate,"%")
```

```python
temp = []
humid = []
for i in range(0,len(test_X)):
    temp.append(test_X.loc[i,"평균기온(°C)"])
    humid.append(test_X.loc[i,"평균 상대습도(%)"])
```

```python
rate = []
for i in range(0,len(test_Y)):
    rate.append(test_Y.loc[i,"증감율"])
```

```python
import numpy as np
import matplotlib.pyplot as plt
# graph of temperature, humid, 
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(train_x, train_y, train_z, marker='o', s=15, cmap='Greens')
plt.show()
```

```python
fig = plt.figure(figsize=(6, 6)) # real data
ax = fig.add_subplot(111, projection='3d')
ax.scatter(temp, humid, rate, marker='o', s=15, cmap='Greens')
plt.show()
```
